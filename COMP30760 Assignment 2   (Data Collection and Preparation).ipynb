```python
# connect to webpage
r1 = requests.get(TLSA_site)
r1.status_code
# connect to webpage
r2 = requests.get(FB_site)
r2.status_code
# connect to webpage
r3 = requests.get(EA_site)
r3.status_code
# connect to webpage
r4 = requests.get(NFLX_site)
r4.status_code

print(r1,r2,r3,r4)
```

```python
# get the HTML content from the webpage
htmll = r1.content
html2 = r3.content
html3 = r3.content
html4 = r4.content
```

# Extracting tables with Beautiful Soup

```python
# Create the soup
soup1 = BeautifulSoup(htmll, 'lxml')
soup2 = BeautifulSoup(html2, 'lxml')
soup3 = BeautifulSoup(html3, 'lxml')
soup4 = BeautifulSoup(html4, 'lxml')
```

```python
# Tables are marked with the 'table' tag in HTML
tables1 = soup1.find_all("table")
tables2 = soup2.find_all("table")
tables3 = soup3.find_all("table")
tables4 = soup4.find_all("table")
```
